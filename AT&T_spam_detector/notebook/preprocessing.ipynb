```python
import pandas as pd
import plotly.express as px
```

```python
df_spam = pd.read_csv('../datas/spam.csv', encoding='latin-1')
df_spam.info()
```

```python
df_spam.head()
```

```python
mask_unnamed_2 = df_spam["Unnamed: 2"].notnull()
df_spam[mask_unnamed_2].head()
```

```python
mask_unnamed_3 = df_spam["Unnamed: 3"].notnull()
df_spam[mask_unnamed_3].head()
```

```python
mask_unnamed_4 = df_spam["Unnamed: 4"].notnull()
df_spam[mask_unnamed_4].head()
```

It appears that each new "unnamed column" is the follow up of the previous message in column v2, as columns with NaN don't seem to be followed by something other than NaN

```python
df_spam["v2"] = (df_spam["v2"].fillna("") + df_spam["Unnamed: 2"].fillna("") + df_spam["Unnamed: 3"].fillna("") + df_spam["Unnamed: 4"].fillna(""))
to_drop = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"]
df_spam = df_spam.drop(columns=to_drop)
df_spam.head()
```

```python
#Let's rename columns for clarity, label and content message
df_spam = df_spam.rename(columns={"v1": "label_string", "v2": "message"})
df_spam.head()
```

```python
#Adding a column in binary for the label
df_spam["label"] = df_spam["label_string"].map({"ham": 0, "spam": 1})
df_spam.head()
```

```python
fig = px.histogram(df_spam, x="label_string", color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(title="ham/spam repartition", yaxis_title="Nb of messages", xaxis_title="Type of message", title_x=0.5)
fig.show()
```

It appears that roughly 1/6 messages is a spam message, so their repartition is not balanced

```python
df_spam.to_csv("../datas/spam_clean.csv", index=False)
```
